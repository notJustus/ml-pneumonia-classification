```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
israelPalestine_filepath = "data/fatalities_israel_palestine.csv"
israelPalestine_data = pd.read_csv(israelPalestine_filepath)
```

```python
israelPalestine_data.head()
```

```python
sns.scatterplot(x=israelPalestine_data['event_location_region'], y=israelPalestine_data['age'])
```
